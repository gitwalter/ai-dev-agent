````python
from IPython.display import display, Markdown, HTML

def display_groq_response(response, show_tools=True, show_reasoning=False):
    """
    Display Groq LLM response in a nicely formatted way.
    
    Args:
        response: The Groq API response object
        show_tools: Whether to display executed tools summary (default: True)
        show_reasoning: Whether to display the reasoning content (default: False)
    """
    message = response.choices[0].message
    
    # Display main content as Markdown
    if message.content:
        display(Markdown("## LLM Response\n---"))
        display(Markdown(message.content))
    
    # Display executed tools summary
    if show_tools and message.executed_tools:
        display(Markdown("\n---\n## Executed Tools Summary"))
        
        tools_summary = []
        for i, tool in enumerate(message.executed_tools, 1):
            tool_type = tool.type if hasattr(tool, 'type') else 'unknown'
            args = tool.arguments if hasattr(tool, 'arguments') else ''
            
            # Truncate arguments if too long
            if len(args) > 100:
                args_display = args[:100] + "..."
            else:
                args_display = args
            
            tools_summary.append(f"**Tool {i}**: `{tool_type}`\n- Arguments: `{args_display}`")
        
        display(Markdown("\n\n".join(tools_summary)))
        display(Markdown(f"\n**Total tools executed**: {len(message.executed_tools)}"))
    
    # Display reasoning if available and requested
    if show_reasoning and hasattr(message, 'reasoning') and message.reasoning:
        display(Markdown("\n---\n## Reasoning"))
        # Truncate reasoning if too long
        reasoning_text = message.reasoning
        if len(reasoning_text) > 2000:
            reasoning_text = reasoning_text[:2000] + "\n\n... [truncated]"
        display(Markdown(reasoning_text))
    
    # Display usage metadata
    if hasattr(response, 'usage') and response.usage:
        usage = response.usage
        display(Markdown("\n---\n## Usage Statistics"))
        display(Markdown(f"- **Prompt tokens**: {usage.prompt_tokens if hasattr(usage, 'prompt_tokens') else 'N/A'}"))
        display(Markdown(f"- **Completion tokens**: {usage.completion_tokens if hasattr(usage, 'completion_tokens') else 'N/A'}"))
        display(Markdown(f"- **Total tokens**: {usage.total_tokens if hasattr(usage, 'total_tokens') else 'N/A'}"))

print("Display function loaded.")
````

````python
# =============================================================================
# TRACING RAW GROQ CLIENT IN LANGSMITH
# =============================================================================
# 
# LangSmith doesn't have a built-in wrapper for Groq like it does for OpenAI,
# but you can use the @traceable decorator to log any function call.
#
# Method 1: Use @traceable decorator (recommended)
# Method 2: Use OpenTelemetry instrumentation
# =============================================================================

import os
from langsmith import traceable
from groq import Groq

# Ensure LangSmith tracing is enabled
os.environ["LANGSMITH_TRACING"] = "true"
# os.environ["LANGSMITH_API_KEY"] = "your-api-key"  # Set if not already in env
# os.environ["LANGSMITH_PROJECT"] = "groq-tracing-demo"  # Optional: custom project name

client = Groq(api_key=os.environ.get("GROQ_API_KEY"))

@traceable(
    run_type="llm",  # Marks this as an LLM call for proper rendering in LangSmith
    name="groq_compound_chat",  # Custom name for the trace
    metadata={
        "ls_provider": "groq",  # Provider for cost tracking
        "ls_model_name": "groq/compound",  # Model name
    }
)
def traced_groq_chat(messages: list, model: str = "groq/compound") -> dict:
    """
    Traced wrapper for Groq chat completions.
    All calls will be logged to LangSmith.
    """
    response = client.chat.completions.create(
        model=model,
        messages=messages
    )
    
    # Return in a format LangSmith can render nicely
    return {
        "content": response.choices[0].message.content,
        "model": model,
        "executed_tools": [
            {"type": t.type, "arguments": t.arguments} 
            for t in (response.choices[0].message.executed_tools or [])
        ],
        "usage": {
            "prompt_tokens": response.usage.prompt_tokens if response.usage else None,
            "completion_tokens": response.usage.completion_tokens if response.usage else None,
            "total_tokens": response.usage.total_tokens if response.usage else None,
        }
    }

print("[OK] Traced Groq function defined")
````

````python
# Test the traced Groq function - this will appear in LangSmith!

result = traced_groq_chat(
    messages=[
        {"role": "user", "content": "What is the current stock price of Apple?"}
    ],
    model="groq/compound"
)

print("=== Traced Groq Response ===")
print(f"\nContent:\n{result['content'][:500]}...")
print(f"\nTools executed: {len(result['executed_tools'])}")
print(f"Usage: {result['usage']}")
print("\n[OK] Check LangSmith dashboard for the trace!")
````

````python
# =============================================================================
# ADVANCED: Nested tracing with parent chain
# =============================================================================
# You can nest @traceable functions to create a trace tree showing the full
# workflow in LangSmith.

@traceable(name="market_research_workflow")
def market_research(company: str) -> dict:
    """
    Full market research workflow with nested LangSmith tracing.
    This creates a parent trace with the Groq call as a child.
    """
    # This will appear as a child trace under "market_research_workflow"
    result = traced_groq_chat(
        messages=[
            {"role": "system", "content": "You are a financial analyst. Be concise."},
            {"role": "user", "content": f"Give me a brief market analysis of {company}"}
        ],
        model="groq/compound"
    )
    
    return {
        "company": company,
        "analysis": result["content"],
        "tools_used": len(result["executed_tools"])
    }

# Run the workflow - creates nested traces in LangSmith
research = market_research("Tesla")

print("=== Market Research Result ===")
print(f"Company: {research['company']}")
print(f"Tools used: {research['tools_used']}")
print(f"\nAnalysis:\n{research['analysis'][:800]}...")
print("\n[OK] Check LangSmith for nested trace: market_research_workflow -> groq_compound_chat")
````

## LangSmith Tracing for Raw Groq Client - Summary

### Key Concepts:

| Feature | Description |
|---------|-------------|
| `@traceable` | Decorator that logs any function call to LangSmith |
| `run_type="llm"` | Marks the trace as an LLM call for proper UI rendering |
| `ls_provider` | Enables cost tracking by identifying the provider |
| `ls_model_name` | Model name for analytics and filtering |
| Nested traces | Child `@traceable` functions appear as nested spans |
